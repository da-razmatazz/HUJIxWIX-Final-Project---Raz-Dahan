```python
import numpy as np
import pandas as pd
agents = pd.read_csv("Quick_agents_details_data.csv")
interactions = pd.read_excel("Quick_agents_interactions_data_october_december_2021.xlsx")
```

```python
#converting 'start_date' column to datetime due to problems that will come later in the code.
agents['start_date_at_quick'] = pd.to_datetime(agents['start_date_at_quick'])

# checking
agents.info()
```

```python
#converting the agent_id columns in both tables for merging them later.
agents['agent_id'] = agents['agent_id'].astype(str)
interactions['agent_id'] = interactions['agent_id'].astype(str)
```

```python
# merging tables on 'agent_id' column.
joined = pd.merge(agents, interactions, how='left', on='agent_id')
joined.head()
```

```python
# creating a new column of 'work_period_days'.
end_date = pd.to_datetime('2021-12-31')

#work_period
joined['work_period_days'] = (end_date - joined['start_date_at_quick']).dt.days

# reordering the columns for a better view.
new_order=['agent_id', 'agent_type', 'agent_language', 'start_date_at_quick','work_period_days',
           'interaction_id','interaction_date_new','channel','overall_service_ans','nps_ans','nps_reason',
          'is_offer_exist','is_offer_suggested','is_offer_implemented']
joined = joined.reindex(columns=new_order)

joined.head()
```
